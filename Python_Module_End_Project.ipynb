Identify the predominant age group among employees.

```python
data["Age"].unique()
```

```python

bins = [0, 24, 29, 34, 39, np.inf] 
labels = ["20-24", "25-29", "30-34", "35-39", "40+"]

data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
print(data[['Age', 'Age Group']])
```

```python
age_countg=data['Age Group'].value_counts()
age_countg
```

```python
d = pd.DataFrame({'Age Group': age_countg.index, 'Count': age_countg.values})
Age_group_greatcount = age_countg.idxmax()
d = d.sort_values('Age Group')
```

```python
d
```

```python
Age_group_greatcount
```

```python
max_value = d['Count'].max()
colors = ['red' if count == max_value else 'blue' for count in d['Count']]
max_value = d['Count'].max()
plt.figure(figsize=(8, 5))
sns.barplot(x=age_countg.index, y=age_countg.values,hue=age_countg.index, palette=colors)

plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.title("Players Distribution by Age Group", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
